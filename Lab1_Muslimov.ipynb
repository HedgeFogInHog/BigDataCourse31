###4 задание
Вычисление по формуле
\begin{equation*}
f = \frac{9 \pi t + 10 cos(x)}{ \surd t - sin(x)}e^{-2t}
\end{equation* *italicised text*}

```python
import math

def calculate_f(t, x):
    if t <= 0:
        raise ValueError("t должен быть положительным, так как находится под корнем")

    # numerator
    numerator = 9 * math.pi * t + 10 * math.cos(x)

    # denominator
    denominator = math.sqrt(t) - math.sin(x)

    if denominator == 0:
        raise ValueError("Знаменатель не должен быть равен нулю")

    # exponent
    exponent = math.exp(-2 * t)

    # formula
    f = (numerator / denominator) * exponent
    return f
```

```python
# Example
t = 5  # t
x = math.pi / 4  # x

result = calculate_f(t, x)
print(f"Результат: {result}")
```

###5 задание

```python
def find_numbers_in_interval(a, b, num1, num2, num3):
    if a > b:
        raise ValueError("Левая граница интервала должна быть меньше или равна правой границе.")

    numbers = [num1, num2, num3]
    result = [num for num in numbers if a <= num <= b]

    return result
```

```python
# Пример использования
a = 2
b = 5
num1 = 2
num2 = 3
num3 = 6

result = find_numbers_in_interval(a, b, num1, num2, num3)
print(f"Числа в интервале [{a}, {b}]: {result}")
```

```python
def calculate_sequence_sum():
    total_sum = 0
    for n in range(1, 20 + 1):
        # Вычисление текущего элемента последовательности
        element = n / ((n + 1) + (n + 2))
        print(element)
        # Добавление элемента к сумме
        total_sum += element
    return round(total_sum)

print(f"Сумма 20 элементов последовательности: {calculate_sequence_sum()}")
```
